```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Employee-Attrition.csv')
data
```

```python
data = data.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1)
data.columns
```

```python
data.loc[data['Attrition']=='No','Attrition'] = 0
data.loc[data['Attrition']=='Yes','Attrition'] = 1
```

```python
data
```

```python
data['Attrition']
```

```python
data.loc[data['Department']=='Sales','Department']=0
data.loc[data['Department']=='Research & Development','Department']=1
data.loc[data['Department']=='Human Resources','Department']=2
```

```python
data['Department']
```

```python
data.loc[data['Gender']=='Female','Gender']=0
data.loc[data['Gender']=='Male','Gender']=1
```

```python
data['Gender']
```

```python
data.loc[data['OverTime']=='Yes','OverTime']=0
data.loc[data['OverTime']=='No','OverTime']=1
data['OverTime']
```

```python
data.loc[data['EducationField']=='Life Sciences','EducationField']=0
data.loc[data['EducationField']=='Medical','EducationField']=1
data.loc[data['EducationField']=='Marketing','EducationField']=2
data.loc[data['EducationField']=='Technical Degree','EducationField']=3
data.loc[data['EducationField']=='Human Resources','EducationField']=4
data.loc[data['EducationField']=='Other','EducationField']=5
data['EducationField']
```

```python
data.loc[data['JobRole']=='Sales Executive','JobRole']=0
data.loc[data['JobRole']=='Research Scientist','JobRole']=1
data.loc[data['JobRole']=='Laboratory Technician','JobRole']=2
data.loc[data['JobRole']=='Manufacturing Director','JobRole']=3
data.loc[data['JobRole']=='Healthcare Representative','JobRole']=4
data.loc[data['JobRole']=='Manager','JobRole']=5
data.loc[data['JobRole']=='Sales Representative','JobRole']=6
data.loc[data['JobRole']=='Research Director','JobRole']=7
data.loc[data['JobRole']=='Human Resources','JobRole']=8
data['JobRole']
```

```python
data['Attrition'] = data['Attrition'].astype('int')
data['Gender'] = data['Gender'].astype('int')
data['OverTime'] = data['OverTime'].astype('int')
```

```python
data
```

```python
cols=['BusinessTravel','DailyRate','DistanceFromHome','Education', 'EnvironmentSatisfaction','HourlyRate','MaritalStatus','RelationshipSatisfaction','StockOptionLevel','WorkLifeBalance']
data= data.drop(cols,axis=1)
```

```python
data
```

```python
x=data.loc[:,data.columns!='Attrition'].values
x
```

```python
y=data.iloc[:,1:2].values
y
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
